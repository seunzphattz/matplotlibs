### Matplotlib tutorial

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
x_1 = np.linspace(0, 5, 10) # will generate 10 floats from 0 to 5
y_1 = x_1**2
```

### Functional plots

```python
plt.plot(x_1, y_1, "r")

plt.title("Days Square chart")
plt.xlabel("Days")
plt.ylabel(" Days Squared")
```

```python
plt.subplot(1, 2, 1)
plt.plot(x_1, y_1, "r")
```

```python
fig_1 = plt.figure(figsize=(5,4), dpi=100)
axes_1 = fig_1.add_axes([0.1,0.1,0.9,0.9])
axes_1.set_xlabel("Days")
axes_1.set_ylabel("Days squared")
axes_1.set_title("days squared chart")
axes_1.plot(x_1, y_1, label="x/x**2")
axes_1.plot(y_1,x_1, label= "x**2/x")
```

```python
data = {"Temperature": [37, 40, 42, 49, 53],
       "Sales": [292, 228, 320, 324, 324]}

df = pd.DataFrame(data)
np_arr = df.values

x_1= np_arr[:, 0]
y_1 =np_arr[:, 1]


fig_1 = plt.figure(figsize=(6, 4))
axes_1 = fig_1.add_axes([0, 0, 1, 1])
axes_1.set_title("Temperature vs Sales")
axes_1.set_xlabel("Temperature")
axes_1.set_ylabel("Sales")
axes_1.plot(x_1, y_1, label = "Temp/sales")
axes_1.legend(loc= 0)


# annotate text to a point on the axes


axes_1.annotate("Good Month", xy= (52,320), xytext=(44, 311), arrowprops = dict(facecolor="black", shrink= 0.05))

plt.bar(x_1, y_1) # bar chart notation  

axes_1.text(44, 280, r'$\alpha$')
```

```python
import datetime 
df = pd.read_csv("GOOG.csv")
goog_data = df.to_numpy()
goog_arr = goog_data[:, 4]




holidays = [datetime.datetime(2020,5,25), datetime.datetime(2020, 8, 19)]

date_arr = pd.bdate_range(start= '5/20/2020', end='8/19/2020', freq ="C", holidays= holidays)

date_arr_np= date_arr.to_numpy()

# date_arr_np







# fig_7= plt.figure(figsize =(8, 5))
# axes_7 = fig_7.add_axes([0.1,0.1, 0.9,0.9])
# plt.plot(date_arr_np, goog_arr)

df.to_numpy()[:, 4]
stk_data = df[-5:]

col_head = ("Date", "Open", "High", "low", "Close", "Adj Close", "Volume")

stk_data_np= stk_data.to_numpy()

plt.figure(linewidth = 2, tight_layout={"pad":5}, figsize=(5,3))

axes_9 = plt.gca()

axes_9.get_xaxis().set_visible(False)
axes_9.get_yaxis().set_visible(False)
plt.box(on=None)


ccolors= plt.cm.Blues(np.full(len(col_head), 0.2))

the_table = plt.table(cellText=stk_data_np, loc="center", colLabels = col_head, colColours = ccolors)

the_table.set_fontsize(14)
the_table.scale(3, 2.5)

```
